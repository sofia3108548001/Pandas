```python
import geopandas as gpd
import pandas as pd

!pip install contextily
import contextily as cx
```

```python
localidades = gpd.read_file("localidades_bogota.json")
localidades
```

```python
localidades.plot()
```

```python
Población =pd.read_csv("poblacion_bogota.csv", encoding='latin1', delimiter=";")
Población
```

```python
Población_24 =Población[Población['Año'] == 2024]
Población_24
```

```python
grupos= Población_24.groupby("Localidad")
print(grupos)
```

```python
totalPoblacionLocaliad= grupos["Población"].sum()
print(totalPoblacionLocaliad)
```

```python
dfPoblacion= totalPoblacionLocaliad.reset_index()
dfPoblacion
```

```python
dfPoblacion["Localidad"]= dfPoblacion["Localidad"].str.upper()

dfPoblacion
```

# Quitar las tildes de la columna Localidad

```python
dfPoblacion['Localidad']

def replace_accents(input_str):
  replacements ={
      'Á': 'A',
      'É': 'E',
      'Í': 'I',
      'Ó': 'O',
      'Ú': 'U',
  }

  for accented_char, unaccented_char in replacements.items():
    input_str = input_str.replace(accented_char, unaccented_char)
  return input_str

dfPoblacion['Localidad'] = dfPoblacion['Localidad'].apply(replace_accents)
```

```python
dfPoblacion
```

```python
filas= dfPoblacion["Localidad"]== "LA CANDELARIA"
filas
```

```python
columna= 'Localidad'
dfPoblacion.loc[filas,columna]= "CANDELARIA"
dfPoblacion
```

```python
merge = localidades.merge(dfPoblacion, left_on= "LocNombre", right_on= "Localidad")

merge
```

```python
merge.drop("Localidad", axis=1, inplace= True)
merge
```

```python
mapa_2= merge.plot(
    column= 'Población',
    figsize=(15,15),
    edgecolor='black',
    cmap= 'tab20',
    legend= True,

)

cx.add_basemap(mapa_2, crs=merge.crs.to_string())
```
