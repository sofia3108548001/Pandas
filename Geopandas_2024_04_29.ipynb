#  **Solución de la tarea **


```python
# Calcular la distancia del centroide de las localidades hasta la Universidad Distrital Facultad de ingenieria.

from shapely.geometry import Point
Universidad= Point (-74.06570148805335, 4.628133695217806)
Universidad
```

```python
#point (Longuitud, Latitud)
Universidad = gpd.GeoSeries (Universidad, crs=4326)
Universidad
```

```python
Universidad= Universidad.to_crs(epsg=9377)
Universidad
```

```python
#tomar la distancia desde la UD a cada uno de los puntos del centroide
gdf_localidades["distancia_u"]=[Universidad.distance (centroid) for centroid in gdf_localidades.centroide]
gdf_localidades["distancia_u"]
```
